# Naive Spatial Compression 

Go through data and reduce resolution



### target resolution 2: 54x30

```python
import os
import subprocess

resolution = (107,60)

DATA_SOURCE = "data_right_left_up_down_1200"
DATA_TARGET = f"data_{resolution[1]}_right_left_up_down_1200"

if DATA_TARGET not in os.listdir():
    os.mkdir(DATA_TARGET)
    os.mkdir(DATA_TARGET + "/test")
    os.mkdir(DATA_TARGET + "/train")
    os.mkdir(DATA_TARGET + "/validation")
    status = subprocess.call(f'cp {DATA_SOURCE + "/indx_df.csv"} {DATA_TARGET + "/indx_df.csv"}', shell=True) 
```

```python
def conver_files_in_folder(source, destination, resolution):
    for file in os.listdir(source):
        subprocess.call(
            f'ffmpeg -i {source + "/" + file} -vf scale={resolution[0]}x{resolution[1]} -preset slow -crf 18 {destination + "/" + file}',
            shell=True
        )

conver_files_in_folder(DATA_SOURCE + "/test", DATA_TARGET + "/test", (213, 120))
conver_files_in_folder(DATA_SOURCE + "/train", DATA_TARGET + "/train", (213, 120))
conver_files_in_folder(DATA_SOURCE + "/validation", DATA_TARGET + "/validation", (213, 120))
```
